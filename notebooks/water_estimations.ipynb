```python
import pandas as pd
import numpy as np
```

```python
df_project = pd.read_csv('../data/clean/df_leed_cleaned.csv')
df_project
```

```python
df_project['City'] = df_project['City'].fillna('NOA')
df_project
```

```python
df_project.loc[df_project['ProjectName'] == 'CONF', 'State'] = 'CON'
df_project.loc[(df_project['City'].isin(['SHA', 'BEI', 'TIA','SHA', 'HOR', 'HOH', 'CHA', 'CD', 'XNA', 'UC', 'HAE', 'HAR', 'SHI', 'SHE'])) & (df_project['State'].isna()), 'State'] = df_project['City']
df_project
```

```python
df_project.copy()
```

```python
#Estimate water use intensity 
water_use_intensity = {
    580: ['Public Order and Safety: Fire/Police Station', 'Office: Government', 'Office: Administrative/Professional',
          'Multi-Family Residential: Apartment', 'Public Assembly: Other Assembly', 'Multi-Family Residential: Condominium', 
          'Public Assembly: Museum', 'Public Assembly: Entertainment', 'Core Learning Space: Preschool/Daycare', 
          'Core Learning Space: K-12 High School', 'Office: Mixed-Use', 'Service: Vehicle Storage/Maintenance', 
          'Office: Other Office', 'Public Assembly: Social/Meeting', 'Public Assembly: Recreation',
          'Multi-Family Residential: Lowrise', 'Public Assembly: Convention Center', 'Office: Financial'],

    220: ['Retail: Grocery Store/Food Market', 'Retail: Open Shopping Center', 'Retail: Convenience Store', 
          'Retail: Restaurant/Cafeteria', 'Retail: Bank Branch'],

    150: ['CONF','Full Service', 'Other', 'Service: Other Service', 'Public Order and Safety: Other Public Order',
          'Religious Worship', 'Service: Vehicle Service/Repair'],
    
    500:  ['Core Learning Space: College/University', 'Public Assembly: Library', 'Core Learning Space: K-12 Elementary/Middle School', 
           'Core Learning Space: Other classroom education'],

    2200: ['Lodging: Hotel/Motel/Resort', 'Lodging: Hotel/Motel/Resort, Limited Service', 'Lodging: Dormitory', 
          'Retail: Enclosed Mall', 'Retail: Other Retail', 'Retail: Fast Food', 'Lodging: Other lodging', 'Lodging: Hotel/Motel/Resort, Full Service'],
   
    150: ['Warehouse: General', 'Warehouse: Self Storage Units', 'Warehouse: Nonrefrigerated Distribution/Shipping',
          'Warehouse: Refrigerated', 'Data Center'],

    2800: ['Health Care', 'Industrial Manufacturing', 'Healthcare: Nursing Home/ Assisted Living',
          'Airport: Distribution Center', 'Healthcare: Inpatient', 'Healthcare: Clinic/Other Outpatient',
          'Public Assembly: Stadium/Arena', 'Airport: Other', 'Airport: Public Order/Fire/Police', 'Laboratory', 
          'Healthcare: Outpatient Office (Diagnostic)', 'Airport: Terminal/Concourse', 'Airport: Hangar', 
          'Transit: Station/Elevated', 'Transit: Other', 'Airport: Office', 'Office: Medical (Non-Diagnostic)', 
          'Transit: Station/Underground', 'Transit: Maintenance/Storage']}

wui_map = {}
for wui, types in water_use_intensity.items():
    for t in types:
        wui_map[t] = wui

df_project['water_use_intensity'] = df_project['ProjectTypes'].map(wui_map)

df_project['water_use_intensity'] = df_project['water_use_intensity'].fillna(500)  # or another default
df_project
```

```python
df_project = df_project.dropna()
df_project.info()
```

```python
leed_water_saving = {
    "Certified": 10,   
    "Silver": 20,      
    "Gold": 30,        
    "Platinum": 40}

df_project['Water_Saving_%'] = df_project['CertLevel'].map(leed_water_saving)

#baseline water use (liters/year)
df_project['Water_Baseline_L'] = df_project['GrossFloorAreaSqm'] * df_project['water_use_intensity']

#water savings (liters/year)
df_project['Water_Saving_L'] = df_project['Water_Baseline_L'] * (df_project['Water_Saving_%'] / 100)

#liters to cubic meters
df_project['Water_Saving_m3'] = df_project['Water_Saving_L'] / 1000

df_project['Water_Baseline_L'] = df_project['Water_Baseline_L'].round()
df_project['Water_Saving_L'] = df_project['Water_Saving_L'].round()
df_project['Water_Saving_m3'] = df_project['Water_Saving_m3'].round(3)  

water_estimates_df = df_project[['ID', 'ProjectName', 'Country', 'CertLevel', 'CertDate', 'GrossFloorAreaSqm', 'water_use_intensity', 'Water_Saving_%', 'Water_Baseline_L',
                               'Water_Saving_L', 'Water_Saving_m3']]

water_estimates_df
```

```python
water_estimates_df = water_estimates_df.rename(columns={'ID': 'Project_Id'})
water_estimates_df
```

```python
water_estimates_df.to_csv('water_savings.csv', index=False)
```
