```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("../data/clean/df_leed_cleaned.csv")
df.head()
```

```python
df = df.rename(columns={'ID': 'Project_Id'})
df
```

```python
df.isnull().sum()
```

```python
for index, row in df[df['City'].isna()].iterrows():
    print(f"Row {index}: {row.to_dict()}")
```

```python
df['City'] = df['City'].fillna('NOA')
df.info()
```

```python
df.loc[df['ProjectName'] == 'CONF', 'State'] = 'CON'
df.loc[(df['City'].isin(['SHA', 'BEI', 'TIA','SHA', 'HOR', 'HOH', 'CHA', 'CD', 'XNA', 'UC', 'HAE', 'HAR', 'SHI', 'SHE'])) & (df['State'].isna()), 'State'] = df['City']
df
```

```python
#Create Location Table (filter columns)
location_df = df[['City', 'State', 'Zipcode', 'Country']].copy()

start_number = 1
location_df['Location_Id'] = pd.Series(range(start_number, start_number + len(location_df)))

location_df = location_df[['Location_Id', 'City', 'State', 'Zipcode', 'Country']]
location_df['Zipcode'] = location_df['Zipcode'].astype('object')
location_df
```

```python
location_df.to_csv('location.csv', index=False)
```

```python
#Create Project Table (filter column)
project_df = df[['Project_Id', 'ProjectName', 'GrossFloorAreaSqm', 'ProjectTypes', 'IsCertified']].copy()
project_df
```

```python
project_df = pd.concat([project_df, location_df], axis=1)
columns_to_drop = ['City', 'State', 'Zipcode', 'Country']
df_merged_cleaned = project_df.drop(columns=columns_to_drop)
df_merged_cleaned = df_merged_cleaned[['Project_Id', 'Location_Id', 'ProjectName', 'GrossFloorAreaSqm', 'ProjectTypes', 'IsCertified']]
df_merged_cleaned
```

```python
df_merged_cleaned.info()
```

```python
df_merged_cleaned.to_csv('projects.csv', index=False)
```

```python
#Create Certification Table
cert_df = df[['Project_Id', 'RatingSystem', 'PointsAchieved', 'CertLevel', 'CertDate']].copy()
start_number = 1
cert_df['Certification_Id'] = pd.Series(range(start_number, start_number + len(cert_df)))
cert_df = cert_df[['Certification_Id', 'Project_Id', 'RatingSystem', 'PointsAchieved', 'CertLevel', 'CertDate']]
cert_df['CertDate'] = pd.to_datetime(cert_df['CertDate'], dayfirst=True)
cert_df
```

```python
cert_df.to_csv('certification.csv', index=False)
```
