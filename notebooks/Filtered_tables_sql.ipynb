```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("../data/clean/df_leed_cleaned.csv")
df.head()
```

```python
df = df.rename(columns={'ID': 'Project_Id'})
df
```

```python
df.isnull().sum()
```

```python
for index, row in df[df['City'].isna()].iterrows():
    print(f"Row {index}: {row.to_dict()}")
```

```python
df['City'] = df['City'].fillna('NOA')
df.info()
```

```python
df.loc[df['ProjectName'] == 'CONF', 'State'] = 'CON'
df.loc[(df['City'].isin(['SHA', 'BEI', 'TIA','SHA', 'HOR', 'HOH', 'CHA', 'CD', 'XNA', 'UC', 'HAE', 'HAR', 'SHI', 'SHE'])) & (df['State'].isna()), 'State'] = df['City']
df
```

```python
#Create Location Table (filter columns)
location_df = df[['City', 'State', 'Zipcode', 'Country']].copy()

start_number = 1
location_df['Location_Id'] = pd.Series(range(start_number, start_number + len(location_df)))

location_df = location_df[['Location_Id', 'City', 'State', 'Zipcode', 'Country']]
location_df['Zipcode'] = location_df['Zipcode'].astype('object')
location_df
```

```python
location_df.to_csv('df_location_cleaned.csv', index=False)
```

```python
loc_df= pd.read_csv('../data/clean/df_location_cleaned.csv')
loc_df
```

```python
#Create new Location table unique_countries
#create Country_Id with unique values
country_df = loc_df[['Country']].drop_duplicates().reset_index(drop=True)
country_df['Country_Id'] = country_df.index + 1

loc_df = loc_df.merge(country_df, on='Country', how='left')

#create State_Id with unique values
state_df = loc_df[['State', 'Country_Id']].drop_duplicates().reset_index(drop=True)
state_df['State_Id'] = state_df.index + 1

loc_df = loc_df.merge(state_df, on=['State', 'Country_Id'], how='left')

#create City_Id with unique values
city_df = loc_df[['City', 'State_Id', 'Country_Id']].drop_duplicates().reset_index(drop=True)
city_df['City_Id'] = city_df.index + 1

loc_df = loc_df.merge(city_df, on=['City', 'State_Id', 'Country_Id'], how='left')

#create main location_Id table with conscutive values
loc_df['Location_Id'] = range(1, len(loc_df) + 1)
loc_df
```

```python
loc_df.rename(columns={"City": "CityName", "State": "StateName", "Country": "CountryName"}, inplace=True)
```

```python
#Filtered country table
unique_countries = loc_df[['Country_Id', 'CountryName']].drop_duplicates().reset_index(drop=True)
unique_countries
```

```python
unique_countries.to_csv('country.csv', index=False)
```

```python
#Filtered city table
unique_cities = loc_df[['City_Id', 'CityName', 'State_Id']].drop_duplicates().reset_index(drop=True)
unique_cities = unique_cities[['City_Id', 'State_Id', 'CityName']]
unique_cities
```

```python
unique_cities.to_csv('city.csv', index=False)
```

```python
#Filtered state table
unique_state = loc_df[['State_Id', 'Country_Id', 'StateName']].drop_duplicates().reset_index(drop=True)
#unique_state = unique_state[['City_Id', 'State_Id', 'CityName']]
unique_state
```

```python
unique_state.to_csv('state.csv', index=False)
```

```python
#Filtered location table
filtered_location = loc_df[['Location_Id', 'City_Id', 'Zipcode']].copy()
filtered_location['Zipcode'] = loc_df['Zipcode'].astype(str)
filtered_location
```

```python
filtered_location.to_csv('location.csv', index=False)
```

```python
#Create Project Table (filter column)
project_df = df[['Project_Id', 'ProjectName', 'GrossFloorAreaSqm', 'ProjectTypes', 'IsCertified']].copy()
project_df
```

```python
project_df = pd.concat([project_df, location_df], axis=1)
columns_to_drop = ['City', 'State', 'Zipcode', 'Country']
df_merged_cleaned = project_df.drop(columns=columns_to_drop)
df_merged_cleaned = df_merged_cleaned[['Project_Id', 'Location_Id', 'ProjectName', 'GrossFloorAreaSqm', 'ProjectTypes', 'IsCertified']]
df_merged_cleaned
```

```python
df_merged_cleaned.info()
```

```python
df_merged_cleaned.to_csv('projects.csv', index=False)
```

```python
#Create Certification Table
cert_df = df[['Project_Id', 'RatingSystem', 'PointsAchieved', 'CertLevel', 'CertDate']].copy()
start_number = 1
cert_df['Certification_Id'] = pd.Series(range(start_number, start_number + len(cert_df)))
cert_df = cert_df[['Certification_Id', 'Project_Id', 'RatingSystem', 'PointsAchieved', 'CertLevel', 'CertDate']]
cert_df['CertDate'] = pd.to_datetime(cert_df['CertDate'], dayfirst=True)
cert_df
```

```python
cert_df.to_csv('certification.csv', index=False)
```

```python
#Create Sustaintability Table
df_energy = pd.read_csv('../data/clean/energy_savings.csv')
df_energy
```

```python
df_energy = df_energy.drop(columns=['ProjectName', 'Country', 'CertLevel', 'CertDate', 'GrossFloorAreaSqm'])
df_energy
```

```python
df_sustainability = df_energy.rename(columns={'Energy_Saving_kWh': 'EnergySaving', 'CO2_Avoided_kg': 'CO2Avoided'})
df_sustainability['Metric_Id'] = range(1, len(df_sustainability) + 1)
df_sustainability
```

```python
df_sustainability = df_sustainability[['Metric_Id', 'Project_Id', 'EnergySaving', 'CO2Avoided']]
df_sustainability
```

```python
df_water = pd.read_csv('../data/clean/water_savings.csv')
df_water
```

```python
df_water = df_water.drop(columns=['ProjectName', 'Country', 'CertLevel', 'CertDate', 'GrossFloorAreaSqm',	'water_use_intensity', 'Water_Saving_%', 'Water_Baseline_L', 'Water_Saving_L'])
df_water
```

```python
merged_df = pd.merge(df_sustainability, df_water, on='Project_Id', how='outer')
merged_df= merged_df.rename(columns={'Water_Saving_m3': 'WaterSaving'})
merged_df
```

```python
df_waste = pd.read_csv('../data/clean/waste_savings.csv')
df_waste
```

```python
df_waste = df_waste.drop(columns=['ProjectName', 'City', 'State', 'Zipcode', 'Country', 'RatingSystem',	'PointsAchieved', 'CertLevel', 'CertDate', 'IsCertified', 'GrossFloorAreaSqm', 'ProjectTypes', 'waste_intensity', 'Waste Reduction %', 'Estimated Waste'])
df_waste
```

```python
merged_final_df = pd.merge(merged_df, df_waste, on='Project_Id', how='outer')
merged_final_df= merged_final_df.rename(columns={'Waste Savings': 'WasteAvoided'})
merged_final_df
```

```python
merged_final_df.to_csv('sustainability_metrics.csv', index=False)
```
