```python
countries_to_keep = ['Brazil', 'Canada', 'China', 'Hong Kong', 'India', 'Italy', 'South Korea', 'Mexico', 'Spain', 'Turkey', 'United Arab Emirates', 'United States']  
df_energy = df_energy[df_energy['country'].isin(countries_to_keep)].copy()
df_energy
```

```python
country_abbrev = {
    'Brazil': 'BR',
    'Canada': 'CA',
    'China': 'CN',
    'Hong Kong': 'HK',
    'India': 'IN',
    'Italy': 'IT',
    'South Korea': 'KR',
    'Mexico': 'MX',
    'Spain': 'ES',
    'Turkey': 'TR',
    'United Arab Emirates': 'AE',
    'United States': 'US'}
df_energy['country'] = df_energy['country'].map(country_abbrev)
df_energy
```

Create a DataFrame for Emission Factors by Country

```python
emission_factors_df  = {
    'country': ['AE', 'BR', 'CA', 'CN', 'HK', 'IN', 'IT', 'KR', 'MX', 'TR', 'US', 'ES'],
    'emission_factor_per_country': [118.4, 37.9, 90.2, 163.2, 169.2, 208.9, 72.3, 120.5, 87.9, 124.8, 127.6, 68.0]}

emission_factors_df = pd.DataFrame(emission_factors_df)
emission_factors_df
```

```python
merged_df = pd.merge(emission_factors_df, df_energy, on='country', how='left')
merged_df
```

```python
merged_df = merged_df[['country', 'year', 'population', 'energy_per_capita', 'emission_factor_per_country', 'primary_energy_consumption']]
merged_df
```

```python
merged_df.columns = merged_df.columns.str.capitalize()
merged_df
```

```python
merged_df.to_csv('energy_consumption.csv', index=False)
```
