```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('../data/clean/energy_consumption.csv')
df_1
```

```python
df_2 = pd.read_csv('../data/clean/df_leed_cleaned.csv')
df_2
```

```python
merged_df = pd.merge(df_2, df_1, on='Country', how='inner')
merged_df
```

```python
merged_df.info()
```

```python
merged_df['City'] = merged_df['City'].fillna('NOA')
merged_df
```

```python
merged_df.loc[merged_df['ProjectName'] == 'CONF', 'State'] = 'CON'
merged_df.loc[(merged_df['City'].isin(['SHA', 'BEI', 'TIA','SHA', 'HOR', 'HOH', 'CHA', 'CD', 'XNA', 'UC', 'HAE', 'HAR', 'SHI', 'SHE'])) & (merged_df['State'].isna()), 'State'] = merged_df['City']
merged_df
```

```python
merged_df['Emission_factor_per_country_kgCO2kW'] = (merged_df['Emission_factor_per_country'] / 293.07107).round(2)
merged_df
```

```python
merged_df.copy()
merged_df['ProjectTypes'].unique()
```

```python
energy_intensity_groups = {
    150: ['Public Order and Safety: Fire/Police Station', 'Office: Government', 'Office: Administrative/Professional',
          'Multi-Family Residential: Apartment', 'Public Assembly: Other Assembly', 'Multi-Family Residential: Condominium', 
          'Public Assembly: Museum', 'Public Assembly: Entertainment', 'Core Learning Space: Preschool/Daycare', 
          'Core Learning Space: K-12 High School', 'Office: Mixed-Use', 'Service: Vehicle Storage/Maintenance', 
          'Office: Other Office', 'Public Assembly: Social/Meeting', 'Public Assembly: Recreation',
          'Multi-Family Residential: Lowrise', 'Public Assembly: Convention Center', 'Office: Financial'],

    200: ['Retail: Grocery Store/Food Market', 'Retail: Open Shopping Center', 'Retail: Convenience Store', 
          'Retail: Restaurant/Cafeteria', 'Retail: Bank Branch'],

    250: ['CONF','Full Service', 'Other', 'Service: Other Service', 'Public Order and Safety: Other Public Order',
          'Religious Worship', 'Service: Vehicle Service/Repair'],
    
    300:  ['Core Learning Space: College/University', 'Public Assembly: Library', 'Core Learning Space: K-12 Elementary/Middle School', 
           'Core Learning Space: Other classroom education'],

    325: ['Lodging: Hotel/Motel/Resort', 'Lodging: Hotel/Motel/Resort, Limited Service', 'Lodging: Dormitory', 
          'Retail: Enclosed Mall', 'Retail: Other Retail', 'Retail: Fast Food', 'Lodging: Other lodging', 'Lodging: Hotel/Motel/Resort, Full Service'],
   
    500: ['Warehouse: General', 'Warehouse: Self Storage Units', 'Warehouse: Nonrefrigerated Distribution/Shipping',
          'Warehouse: Refrigerated', 'Data Center'],

    700: ['Health Care', 'Industrial Manufacturing', 'Healthcare: Nursing Home/ Assisted Living',
          'Airport: Distribution Center', 'Healthcare: Inpatient', 'Healthcare: Clinic/Other Outpatient',
          'Public Assembly: Stadium/Arena', 'Airport: Other', 'Airport: Public Order/Fire/Police', 'Laboratory', 
          'Healthcare: Outpatient Office (Diagnostic)', 'Airport: Terminal/Concourse', 'Airport: Hangar', 
          'Transit: Station/Elevated', 'Transit: Other', 'Airport: Office', 'Office: Medical (Non-Diagnostic)', 
          'Transit: Station/Underground', 'Transit: Maintenance/Storage']}

reverse_map = {}
for intensity, types in energy_intensity_groups.items():
    for btype in types:
        reverse_map[btype] = intensity

merged_df['Energy_Intensity_kWh_m2'] = merged_df['ProjectTypes'].map(reverse_map)

merged_df['Estimated_Energy_kWh'] = merged_df['GrossFloorAreaSqm'] * merged_df['Energy_Intensity_kWh_m2']

merged_df
```

```python
merged_df = merged_df.drop(columns=['Year', 'Energy_per_capita', 'Emission_factor_per_country'])
merged_df.head()
```

```python
leed_energy_saving = {
    "Certified": 10,
    "Silver": 15,
    "Gold": 25,
    "Platinum": 35
}

# Map the LEED level to energy saving %
merged_df['Energy_Saving_%'] = merged_df['CertLevel'].map(leed_energy_saving)

# Use your new energy intensity per m²/year column directly
merged_df['Energy_Baseline_kWh'] = merged_df['GrossFloorAreaSqm'] * merged_df['Energy_Intensity_kWh_m2']

#Calculate energy savings in kWh
merged_df['Energy_Saving_kWh'] = merged_df['Energy_Baseline_kWh'] * (merged_df['Energy_Saving_%'] / 100)

# Calculate CO2 avoided, using emission factor (kg CO2 / kWh)
merged_df['CO2_Avoided_kg'] = merged_df['Energy_Saving_kWh'] * merged_df['Emission_factor_per_country_kgCO2kW']

# Select columns for the final estimates dataframe
estimates_df = merged_df[['ID', 'ProjectName', 'Country', 'CertLevel', 'CertDate', 'GrossFloorAreaSqm',
                          'Energy_Saving_kWh', 'CO2_Avoided_kg']]

estimates_df.head()
```

```python
estimates_df = estimates_df.dropna()
estimates_df
```

```python
estimates_df = estimates_df.rename(columns={'ID': 'Project_Id'})
estimates_df
```

```python
estimates_df.to_csv('energy_savings.csv', index=False)
```
