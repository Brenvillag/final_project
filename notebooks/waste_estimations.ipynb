```python
import pandas as pd
import numpy as np
```

```python
df_project = pd.read_csv('../data/clean/df_leed_cleaned.csv')
df_project
```

```python
df_project['City'] = df_project['City'].fillna('NOA')
df_project
```

```python
df_project.loc[df_project['ProjectName'] == 'CONF', 'State'] = 'CON'
df_project.loc[(df_project['City'].isin(['SHA', 'BEI', 'TIA','SHA', 'HOR', 'HOH', 'CHA', 'CD', 'XNA', 'UC', 'HAE', 'HAR', 'SHI', 'SHE'])) & (df_project['State'].isna()), 'State'] = df_project['City']
df_project
```

```python
#Estimate waste intensity 
waste_intensity = {
    60: ['Public Order and Safety: Fire/Police Station', 'Office: Government', 'Office: Administrative/Professional',
          'Multi-Family Residential: Apartment', 'Public Assembly: Other Assembly', 'Multi-Family Residential: Condominium', 
          'Public Assembly: Museum', 'Public Assembly: Entertainment', 'Core Learning Space: Preschool/Daycare', 
          'Core Learning Space: K-12 High School', 'Office: Mixed-Use', 'Service: Vehicle Storage/Maintenance', 
          'Office: Other Office', 'Public Assembly: Social/Meeting', 'Public Assembly: Recreation',
          'Multi-Family Residential: Lowrise', 'Public Assembly: Convention Center', 'Office: Financial'],

    35: ['Retail: Grocery Store/Food Market', 'Retail: Open Shopping Center', 'Retail: Convenience Store', 
          'Retail: Restaurant/Cafeteria', 'Retail: Bank Branch'],

    50: ['CONF','Full Service', 'Other', 'Service: Other Service', 'Public Order and Safety: Other Public Order',
          'Religious Worship', 'Service: Vehicle Service/Repair'],
    
    25:  ['Core Learning Space: College/University', 'Public Assembly: Library', 'Core Learning Space: K-12 Elementary/Middle School', 
           'Core Learning Space: Other classroom education'],

    70: ['Lodging: Hotel/Motel/Resort', 'Lodging: Hotel/Motel/Resort, Limited Service', 'Lodging: Dormitory', 
          'Retail: Enclosed Mall', 'Retail: Other Retail', 'Retail: Fast Food', 'Lodging: Other lodging', 'Lodging: Hotel/Motel/Resort, Full Service'],
   
    40: ['Warehouse: General', 'Warehouse: Self Storage Units', 'Warehouse: Nonrefrigerated Distribution/Shipping',
          'Warehouse: Refrigerated', 'Data Center'],

    100: ['Health Care', 'Industrial Manufacturing', 'Healthcare: Nursing Home/ Assisted Living',
          'Airport: Distribution Center', 'Healthcare: Inpatient', 'Healthcare: Clinic/Other Outpatient',
          'Public Assembly: Stadium/Arena', 'Airport: Other', 'Airport: Public Order/Fire/Police', 'Laboratory', 
          'Healthcare: Outpatient Office (Diagnostic)', 'Airport: Terminal/Concourse', 'Airport: Hangar', 
          'Transit: Station/Elevated', 'Transit: Other', 'Airport: Office', 'Office: Medical (Non-Diagnostic)', 
          'Transit: Station/Underground', 'Transit: Maintenance/Storage']}

wi_map = {}
for wi, types in waste_intensity.items():
    for t in types:
        wi_map[t] = wi

df_project['waste_intensity'] = df_project['ProjectTypes'].map(wi_map)

df_project['waste_intensity'] = df_project['waste_intensity'].fillna(50)  
df_project
```

```python
df_waste=df_project
df_waste.head()
```

```python
waste_reduction_map = {
    'Certified': 30,
    'Silver':    40,
    'Gold':      50,
    'Platinum':  60
}

# Step 2: Map the reduction % to each row
df_waste['Waste Reduction %'] = df_waste['CertLevel'].map(waste_reduction_map)

# Step 3: Calculate Estimated Waste (in kg or appropriate unit)
df_waste['Estimated Waste'] = df_waste['GrossFloorAreaSqm'] * df_waste['waste_intensity']

# Step 4: Calculate Waste Savings
df_waste['Waste Savings'] = df_waste['Estimated Waste'] * df_waste['Waste Reduction %']

# Optional: Round for readability
df_waste[['Estimated Waste', 'Waste Savings']] = df_waste[['Estimated Waste', 'Waste Savings']].round(2)

# Display first few rows
df_waste[['ID', 'ProjectName', 'ProjectTypes', 'CertLevel', 'GrossFloorAreaSqm', 'waste_intensity', 'Estimated Waste', 'Waste Savings']]
df_waste.info()
```

```python
df_waste = df_waste.dropna()
df_waste.info()
```

```python
df_waste = df_waste.rename(columns={'ID': 'Project_Id'})
df_waste
```

```python
df_waste.to_csv('waste_savings.csv', index=False)
```
