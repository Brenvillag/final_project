```python
def abbreviate_state(state):
    state = state.strip()
    words = state.split()
    if len(words) == 1:
        return state[:3].upper()
    else:
        return ''.join(word[0].upper() for word in words)

def abbreviate_if_not(state):
    if len(state) == 2 and state.isupper():
        return state  
    else:
        return abbreviate_state(state)

df_leed['State'] = df_leed['State'].apply(abbreviate_if_not)
df_leed
```

```python
def clean_zipcode(z):
    z = str(z).strip()
    if z in ['0', '00000', '0000', '000000']:
        return 0
    if z.upper() == 'CONF':
        return 0
    try:
        return int(z)
    except ValueError:
        return 0  
df_leed['Zipcode'] = df_leed['Zipcode'].apply(clean_zipcode).astype(int)
df_leed.head()
```
